## 3. 폰트 사용하기

설치된 폰트를 사용하는 방법은 3가지가 있습니다. 

    1) `FontProperties` 를 사용하는 방법 - 그래프의 폰트가 필요한 항목마다 지정해 주어야 합니다.
    2) `matplotlib.rcParams[]`으로 전역 글꼴 설정 방법 - 그래프에 설정을 해주면 폰트가 필요한 항목에 적용됩니다.
    3) 2)번의 방법을 `mpl.matplotlib_fname()`로 읽어지는 설정 파일에 직접 적어주는 방법, 단 모든 노트북에 적용됩니다. 노트북을 열 때마다 지정해 주지 않아도 돼서 편리합니다.

### 3.1. 특정 범위만 글꼴 변경하기

원하는 font를 `FontProperties`로 설정하고 특정 부분에만 폰트 설정을 적용합니다.

```python
fontprop = font_manager.FontProperties(fname= '/usr/share/fonts/dejavu/DejaVuSansCondensed-Oblique.ttf', size=18)

plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이', fontproperties=fontprop)
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
plt.show()
```

### 3.2. `matplotlib.rcParams[]으로 전역 글꼴 설정 방법`


```python
# 기본 글꼴로 사용합니다.
mpl.rcParams['font.family'] = 'NanumGothic'
```

```python
data = np.random.randint(-100, 100, 50).cumsum()
plt.plot(range(50), data, 'b')
plt.title('인구 변화')
plt.ylabel('인구수')
plt.xlabel('시간(분)')
```

### 3.3. rcParams 를 설정 파일에 직접 적어주는 방법 - 모든 노트북에 공통적용

설정 파일에 직접 입력합니다.

```python
print ('설정 파일 위치: ', mpl.matplotlib_fname())
```
